```python
from selenium import webdriver
from selenium.webdriver.support import ui
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions
```

```python
URL_HEAD = 'https://okgo.tw'
BROWSER = None
```

```python
def initialize_browser(hidden=True):
    chrome_options = webdriver.ChromeOptions()
    if hidden:
        chrome_options.add_argument('--headless')
    chrome_options.add_argument('log-level=3')
    global BROWSER
    BROWSER = webdriver.Chrome("/usr/lib/chromium-browser/chromedriver", options=chrome_options)
```

```python
mapping = {}

initialize_browser()
BROWSER.get('https://okgo.tw/butywant.html?county=18https://okgo.tw')
#print(BROWSER.page_source)
keyword = BROWSER.find_element_by_css_selector('input.inputkeyword01')
keyword.send_keys('石梯坪')
submit = BROWSER.find_element_by_css_selector('input.input02')
submit.click()
#print(BROWSER.page_source)
result = BROWSER.find_element_by_id('Search_Content_ol')
if result:
    result = result.find_elements_by_tag_name('li')
    mapping['石梯坪'] = []
    for data in result:
        main_window = BROWSER.current_window_handle
        print(data.text)
        title = data.find_element_by_css_selector('div.sItem1').find_element_by_css_selector('a.STopic')
        name = title.get_attribute('title')
        link = title.get_attribute('href')
        desc = data.find_element_by_css_selector('div.sItem2').text
        category = data.find_element_by_css_selector('div.sItem3').text[3:].split('»')
        category = [data.strip() for data in category]
        
        mapping['石梯坪'].append({'name': name, 'desc': desc, 'link': link, 'category': category})
        
for data in mapping['石梯坪']:
    BROWSER.get(data['link'])
    pic_list = BROWSER.find_element_by_css_selector('div.box_content')
    pic_list = pic_list.find_element_by_css_selector('ul.spic')
    pic_list = pic_list.find_elements_by_tag_name('li')
    
    for pic in pic_list:
        pic = pic.find_elements_by_tag_name('a')[0]
        link = pic.get_attribute('href')
        title = pic.get_attribute('title')
        print(link)
        print(title)
    
print(mapping)
```

```python
BROWSER.quit()
```
